```python
import numpy as np
import statsmodels.api as sm
import pylab

test = np.random.normal(20, 5, 1000)

sm.qqplot(test, loc = 20, scale = 5, line = '45')
#pylab.show()
```

1. The points are not normally distributed.
2. The standard error of the mean is an estimation of the standard deviation of the random sampling of means. This is to help arrive at a mean which is representative of the data set rather than of the outliers.
